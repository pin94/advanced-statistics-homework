# Data Visualization

```python
set_gender=list(set(df.Gender))
sort_gender=sorted(set_gender)
for i in range (len(sort_gender)):
  gender=df[df.Gender==sort_gender[i]]
  fig,ax=plt.subplots(1,2,figsize=(15,9))
  fig.suptitle('Customers age with gender: '+str(sort_gender[i]),
               fontsize=20,fontweight='bold',y=1.02)
  gender.Customer_Age.value_counts()[:10].plot(kind='barh',fontsize=12,
                                              ax=ax[0],color='palevioletred')
  gender.Customer_Age.value_counts(ascending=True)[:10].plot(kind='barh',fontsize=12,
                                                            ax=ax[1],color='royalblue')
  ax[0].set_title('Top 10',fontsize=18,fontweight='bold')
  ax[1].set_title('Bottom 10',fontsize=18,fontweight='bold')
  ax[0].set(xlabel='Values',ylabel='Age')
  ax[1].set(xlabel='Values',ylabel='Age')
  fig.tight_layout(pad=3,h_pad=3.0)
```

```python
set_income=list(set(df.Income_Category))
sort_income=sorted(set_income)
for i in range (len(sort_income)):
  income=df[df.Income_Category==sort_income[i]]
  fig,ax=plt.subplots(1,2,figsize=(15,9))
  fig.suptitle('Customers info with income '+str(sort_income[i]),
               fontsize=20,fontweight='bold',y=1.02)
  income.Education_Level.value_counts().plot(kind='barh',fontsize=12,
                                             ax=ax[0],color='mediumseagreen')
  income.Gender.value_counts().plot(kind='barh',fontsize=12,
                                    ax=ax[1],color='slategrey')
  ax[0].set_title('Education',fontsize=18,fontweight='bold')
  ax[1].set_title('Gender',fontsize=18,fontweight='bold')
  ax[0].set(xlabel='Values',ylabel='Education')
  ax[1].set(xlabel='Values',ylabel='Gender')
  fig.tight_layout(pad=3,h_pad=3.0)
```
